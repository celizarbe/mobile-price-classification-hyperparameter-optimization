## Outlier detection

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'whitegrid', palette = 'pastel', color_codes = True)
```

```python
X = raw_df.drop(columns=['price_range']).copy()
y = raw_df.price_range
```

### Univariate analysis

```python
fig, axs = plt.subplots(20, 1, figsize = (10, 40), constrained_layout = True)
fig.suptitle('Columns', fontsize = 15)
for i, column in enumerate(X.columns):
    sns.boxplot(x = X[column], orient = 'h', ax = axs[i])

plt.show()
```

### Bivariate analysis

```python
fig, ax = plt.subplots(figsize = (16, 8))
ax.scatter(X['battery_power'], X['talk_time'])
ax.set_xlabel('Battery power')
ax.set_ylabel('Talk time')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (16, 8))
ax.scatter(X['px_height'], X['px_width'])
ax.set_xlabel('Screen height [px]')
ax.set_ylabel('Screen width [px]')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (16, 8))
ax.scatter(X['int_memory'], X['ram'])
ax.set_xlabel('Memory [GB]')
ax.set_ylabel('RAM [GB]')
plt.show()
```

### Z-score analysis

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(X))
z.head()
```

```python
THRESHOLD = 3
print(np.where(z > THRESHOLD))
```

Column 4 is "fc", which represents a phone's front camera resolution in megapixels. Despite some values being outside the range given a threshold for Z-scores, these data points are not wrong or miscalculated. Removal is not required.
